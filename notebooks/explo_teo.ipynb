```python
import nltk
import pandas as pd

# 1. Téléchargement du corpus Switchboard Dialogue Acts (swda)
nltk.download('swda')
from nltk.corpus import swda

# 2. Extraction des répliques et de leurs intentions (actes de langage)
data = []
# On parcourt un échantillon des dialogues (les 5 premiers pour tester)
for trans in swda.transcript_iter()[:5]: 
    for utterance in trans.utterances:
        # On récupère le texte brut en anglais et le label
        texte = utterance.text
        intention = utterance.act_tag
        data.append({"Texte_brut": texte, "Intention": intention})

# 3. Création d'un DataFrame Pandas pour une belle visualisation
df_explo = pd.DataFrame(data)

# Affichage des 15 premières lignes
print(df_explo.head(15))
```
